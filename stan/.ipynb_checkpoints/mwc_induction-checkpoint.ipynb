# Fitting Dissociation Constants with Stan 

```python
import numpy as np
import bokeh.io 
import bokeh.plotting
import pandas as pd
import pystan
import pickle
import matplotlib.pyplot as plt
import seaborn as sns
bokeh.io.output_notebook()
%matplotlib inline
```

```python
# Load in the data
data = pd.read_csv('../data/flow_master.csv')

# Isolate the single operator and strain we are interested in.
fit_strain = data[(data['rbs']=='RBS1027') & (data['operator']=='O2')]

# Plot the data for sanity.
p = bokeh.plotting.figure(x_axis_label='IPTG [µM]', y_axis_label='fold-change',
                         x_axis_type='log', y_axis_type='log')

p.circle(fit_strain['IPTG_uM'], fit_strain['fold_change_A'], color='dodgerblue')
bokeh.io.show(p)
```

```python
sm = pystan.StanModel(file='fit_kaki.stan')
# Save the compiled statmodel.
with open('fit_kaki.pkl', 'wb') as f:
    pickle.dump(sm, f)
```

```python
dat = {'N': len(fit_strain), 'R':260, 'Nns':4.6E6, 'ep_ai':4.5, 'ep_r':-13.9,
      'n_sites':2, 'c':fit_strain['IPTG_uM'].values,
      'fc':fit_strain['fold_change_A'].values}
fit = sm.sampling(data=dat, iter=4000, chains=4)
```

```python
# Instantiate the canvas.
arrangement = []

# Get all samples without losing chain information
extracted = fit.extract(permuted=False, inc_warmup=False)

# Get the parameter names.
flatnames = fit.flatnames
steps, chains, params = np.shape(extracted)
# Loop through each parameter and chain and plot the steps.
colors = bokeh.palettes.Spectral[11]
for i, f in enumerate(flatnames):
    param_samp = extracted[:, :, i]
    p2 = bokeh.plotting.figure(x_axis_label='steps', y_axis_label='value')
    p1 = bokeh.plotting.figure(x_axis_label='value', y_axis_label='ecdf')
    for j in range(chains):
        p2.line(np.arange(0, steps, 1), param_samp[:, j], alpha=0.5,
                           line_width=1, color=colors[j])
        p2.title.text = f 
        p1.title.text = f 
        
        x, y = np.sort(param_samp[:, j]), np.arange(0, steps, 1) / steps
        p1.line(x, y, color=colors[j])
    arrangement.append(p1)
    arrangement.append(p2)
    
layout = bokeh.layouts.gridplot(arrangement, ncols=2, plot_width=450, plot_height=200)
bokeh.io.show(layout)
```
